```python
import tensorflow as tf
import os 
import cv2
from PIL import Image
import numpy as np 
import matplotlib.pyplot as plt  
```

```python
data = r"C:\computer_vision\animals"
```

```python
imgae_req = ["jpeg","jpg","png","bmp"]
```

```python
os.listdir(data)
```

<h1> Remove Dodgy image</h1>

```python
for image_class in os.listdir(data): 
    image_class_path = os.path.join(data,image_class)
    if not os.path.isdir(image_class_path): 
        continue 
    for images in os.listdir(image_class_path): 
        image_path = os.path.join(image_class_path,images)
        try: 
            img = Image.open(image_path)
            img_type = img.format.lower()
            if img_type not in image_req: 
                print(f"Not under requirment {image_path}")
                os.remove(image_path)
        except Exception as e: 
            print(f"Issue with the image {image_class}: {e}")
```

```python
data = tf.keras.utils.image_dataset_from_directory(data) 
```

```python
data_iterator = data.as_numpy_iterator()
```

```python
batch =  data_iterator.next()
```

```python
batch[1]
```

```python
fig, ax = plt.subplots(ncols = 6, figsize = (20,20))
for idx, img in enumerate(batch[0][:6]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
plt.show()
```

<h1>Pre Processing</h1>

```python
data =  data.map(lambda x,y: (x/255,y))
```

<h1>Split</h1>

```python
len(data)
```

```python
train = int(len(data)*0.7)
test = int(len(data)*0.1)
val = int(len(data)*0.2)+1
```

```python
train+test+val 
```

```python
data_train = data.take(train)
data_val = data.skip(train).take(val)
data_test = data.skip(train + val).take(test)
```

```python
len(data_test)
```

<h1>Model </h1>

```python
from keras.models import Sequential 
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16,(3,3),strides=1, activation = "relu", input_shape = (256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32,(3,3), strides=1, activation = "relu" ))
model.add(MaxPooling2D()) 

model.add(Conv2D(64,(3,3),strides=1,  activation =  "relu" ))
model.add(MaxPooling2D())

model.add(Flatten()) 
model.add(Dense(256, activation = "relu" ))
model.add(Dense(1, activation = "sigmoid"))
```

```python
model.compile("adam", loss = tf.losses.BinaryCrossentropy(),metrics = ["accuracy"])
```

<h1> Train</h1>

```python
hist = model.fit(data_train, epochs = 20, validation_data = data_val)
```

```python
fig = plt.figure()
plt.plot(hist.history["loss"], color = "red", label = "loss" )
plt.plot(hist.history["val_loss"], color = "orange" , label = "val_loss")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history["accuracy"], color = "red" )
plt.plot(hist.history["val_accuracy"], color = "orange")
plt.show()
```

```python
precision = Precision()
recall = Recall()
accuracy = BinaryAccuracy()
```

```python
for batch in data_test.as_numpy_iterator(): 
    x,y = batch
    y_pred = model.predict(x)
    precision.update_state(y,y_pred)
    recall.update_state(y,y_pred)
    accuracy.update_state(y,y_pred)
```

```python
print(precision.result(), recall.result(),accuracy.result() )
```
